```python
import os
import random
import wave
```

```python
folder_list = os.listdir("./Data")

wav_set = []

for folder_name in folder_list:
  wav_list = os.listdir(f"./Data/{folder_name}")
  for wav_name in wav_list:
    wav_set.append((folder_name, wav_name))

random.shuffle(wav_set)

for i, wav in enumerate(wav_set):
  if i == 500: break

  os.makedirs(f"./Data/output/fold{i % 10 + 1}", exist_ok=True)

  with wave.open(f"./Data/{wav[0]}/{wav[1]}", "rb") as wave_read:
    with wave.open(f"./Data/output/fold{i % 10 + 1}/{i % 10 + 1}-{i // 10 + 1}.wav", "wb") as wave_write:
      wave_write.setparams(wave_read.getparams())

      for data in iter(lambda: wave_read.readframes(1024), b''):
        wave_write.writeframes(data)
```
